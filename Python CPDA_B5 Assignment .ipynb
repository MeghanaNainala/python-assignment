```python

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())


categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])
```

```python
duplicates = df.duplicated().sum()
print("Number of duplicate rows:", duplicates)


df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)
```

### Task 3.3  Average Selling Price by Odometer
Odometer readings are grouped into ranges to make the relationship clearer.  
A scatter plot is used to show how price changes with mileage.


```python
# Task 3.3  Average selling price by odometer range

import matplotlib.pyplot as plt
import pandas as pd


bins = [0, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000, 150000, 200000, 300000]
labels = ["0-10k", "10-20k", "20-30k", "30-40k", "40-50k", "50-60k", "60-70k",
          "70-80k", "80-90k", "90-100k", "100-150k", "150-200k", "200-300k"]

df["odometer_bin"] = pd.cut(df["odometer"], bins=bins, labels=labels)


avg_price_by_odometer = df.groupby("odometer_bin")["sellingprice"].mean()

print("Average Selling Price by Odometer Range:")
print(avg_price_by_odometer)


plt.figure(figsize=(10,6))
plt.scatter(avg_price_by_odometer.index.astype(str), avg_price_by_odometer.values)

plt.title("Average Selling Price by Odometer Range")
plt.xlabel("Odometer Range")
plt.ylabel("Average Selling Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Insight
There is a clear downward trend  as odometer increases, selling price decreases.  
High-mileage cars have significantly lower prices due to wear and tear.

### Task 3.5  Average Selling Price by Condition Score Ranges (Size 5)
The condition scores are grouped into bins of size 5.  
For each range, we calculate the average selling price and plot it using a bar chart.
This helps us understand how car condition affects pricing.


```python
# Task 3.5  Average selling price by condition score ranges of size 5

import pandas as pd
import matplotlib.pyplot as plt


bins = list(range(0, 101, 5))  # 0 to 100 with step 5
labels = [f"{i}-{i+5}" for i in range(0, 100, 5)]

df['condition_range'] = pd.cut(df['condition'], bins=bins, labels=labels, right=False)

avg_price_by_condition = df.groupby('condition_range')['sellingprice'].mean()

print("Average Selling Price by Condition Range:")
print(avg_price_by_condition)

# Plot
plt.figure(figsize=(12,6))
avg_price_by_condition.plot(kind='bar')

plt.title("Average Selling Price by Condition Score Ranges (Size 5)")
plt.xlabel("Condition Range")
plt.ylabel("Average Selling Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Insight
The bar chart shows that cars with higher condition scores generally have higher average selling prices.  
There is a clear upward trend as the condition improves.  
This indicates that buyers are willing to pay more for cars in better condition, and condition score is a strong predictor of price.


### Task 3.6  Number of Cars Sold by Condition Score Ranges (Size 10)
Condition scores are divided into bins of size 10 (010, 1020, etc.).  
We count the number of cars in each range and plot a bar chart to understand how car conditions are distributed in the dataset.


```python
# Task 3.6  Number of cars sold by condition ranges of size 10

import pandas as pd
import matplotlib.pyplot as plt


bins = list(range(0, 101, 10))  # 0,10,20,...100
labels = [f"{i}-{i+10}" for i in range(0, 100, 10)]

df['condition_range_10'] = pd.cut(df['condition'], bins=bins, labels=labels, right=False)


cars_count_by_condition = df['condition_range_10'].value_counts().sort_index()

print("Number of Cars in Each Condition Range:")
print(cars_count_by_condition)

# Plot
plt.figure(figsize=(12,6))
cars_count_by_condition.plot(kind='bar')

plt.title("Number of Cars Sold by Condition Score Ranges (Size 10)")
plt.xlabel("Condition Range")
plt.ylabel("Number of Cars Sold")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Insight
The bar chart shows how car conditions are distributed in the dataset.  
Some condition ranges have far more cars than others, indicating common condition scores for used cars.  
Usually, mid-to-high condition ranges have the most cars, showing that dealers list cars in relatively good condition.


### Task 3.7  Box Plot of Selling Price by Color
This box plot shows how selling prices vary across different car colors.  
It helps identify the spread, median price, and presence of outliers for each color.


```python
# Task 3.7  Box plot for selling price grouped by color

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 6))
sns.boxplot(x='color', y='sellingprice', data=df)

plt.title("Selling Price Distribution by Car Color (with Outliers)")
plt.xlabel("Car Color")
plt.ylabel("Selling Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Insight (Before Removing Outliers)
Some colors show very high outliers, meaning a few extremely expensive cars distort the scale.  
The median price and spread differ across colors, but the outliers make it harder to compare the distributions.


```python
# Removing outliers using IQR method

Q1 = df['sellingprice'].quantile(0.25)
Q3 = df['sellingprice'].quantile(0.75)
IQR = Q3 - Q1

# Filter without outliers
df_no_outliers = df[(df['sellingprice'] >= Q1 - 1.5 * IQR) &
                    (df['sellingprice'] <= Q3 + 1.5 * IQR)]

df_no_outliers.shape
```

```python
plt.figure(figsize=(14, 6))
sns.boxplot(x='color', y='sellingprice', data=df_no_outliers)

plt.title("Selling Price Distribution by Car Color (Without Outliers)")
plt.xlabel("Car Color")
plt.ylabel("Selling Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Insight (After Removing Outliers)
After removing outliers, the box plots become clearer and easier to compare.  
We can now observe the true distribution of prices for each color without extreme values affecting the scale.  
This helps identify which colors tend to have higher or lower typical selling prices.
